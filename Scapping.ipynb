```python
def extract_image_url(url):
    
    response = requests.get(url)
     
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser') 
        
        list_of_image_url = []
        image_tag = soup.find_all('img') 
        for img_tag in image_tag:
            if img_tag:
                image_url = img_tag.get('src')
                list_of_image_url.append(image_url)
            else:
                print("No image found on the page.")   
        return list_of_image_url
    else:
        print("Error in link")
        
def download_image(url, save_path):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"Image downloaded successfully to {save_path}")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading image: {e}")
        
 
    
    
formal_img = []
for url in urls:
    l1 = extract_image_url(url)
    formal_img.extend(l1)
    
# duplicated removal 
unique_set = set(formal_img)
unique_formal_img = list(unique_set)
len(unique_formal_img)
    
    
i = 1
for image_url in unique_formal_img:
    save_path = "Dataset/Formal/formal_{}.jpg".format(i)
    download_image(image_url, save_path)
    i+=1
 
```

```python
def extract_image_url(url):
    
    response = requests.get(url)
     
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser') 
        
        list_of_image_url = []
        image_tag = soup.find_all('img') 

        for img_tag in image_tag:
            if img_tag:
                image_url = img_tag.get('src')
                list_of_image_url.append(image_url)
            else:
                print("No image found on the page.")   
        return list_of_image_url
    else:
        print("Error in link")
        
informal_img = []
for url in informal_urls:
    l1 = extract_image_url(url)
    informal_img.extend(l1)
```
